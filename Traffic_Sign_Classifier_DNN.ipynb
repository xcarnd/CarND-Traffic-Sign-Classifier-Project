````python
### Preprocess the data here. Preprocessing steps could include normalization, converting to grayscale, etc.
### Feel free to use as many code cells as needed.

# state for the input training set
is_reshaped = False
is_ohe = False
is_normalized = False
is_randomized = False
````

````python
# convert rgb to grayscale?
grayscaling_samples = True

import cv2

if grayscaling_samples:
    # rgb to grayscale
    print("Using grayscale")
    X_train_input = np.array(list(map(lambda img: cv2.cvtColor(img, cv2.COLOR_RGB2GRAY), X_train)))
    X_valid_input = np.array(list(map(lambda img: cv2.cvtColor(img, cv2.COLOR_RGB2GRAY), X_valid)))
    X_test_input = np.array(list(map(lambda img: cv2.cvtColor(img, cv2.COLOR_RGB2GRAY), X_test)))
else:
    # no preprocessing
    print("Using rgb")
    X_train_input = X_train
    X_valid_input = X_valid
    X_test_input = X_test
    
print("Preprocessing done.")
````

````python
# always randomizing the input data set or randomize only once (checking against is_randomized)
always_randomizing = False
# join train set and valid set then resplit?
join_then_resplit = False

from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle

def normalize(input_minmax, output_minmax, val):
    return val / (input_minmax[1] - input_minmax[0]) * (output_minmax[1] - output_minmax[0]) + output_minmax[0]

if not is_reshaped:
    # reshape input features to row vectors
    X_train_input = np.reshape(X_train_input, (np.shape(X_train_input)[0], -1))
    X_valid_input = np.reshape(X_valid_input, (np.shape(X_valid_input)[0], -1))
    X_test_input = np.reshape(X_test_input, (np.shape(X_test_input)[0], -1))
    print("Reshaping done.")
    is_reshaped = True

if not is_ohe:
    # one-hot encoding for outputs
    encoder = LabelBinarizer()
    encoder.fit(range(n_classes))
    y_train_output = encoder.transform(y_train)
    y_train_output = y_train_output.astype(np.float32)
    y_valid_output = encoder.transform(y_valid)
    y_valid_output = y_valid_output.astype(np.float32)
    y_test_output = encoder.transform(y_test)
    y_test_output = y_test_output.astype(np.float32)
    print("One-Hot-Encoding done.")
    is_ohe = True

if not is_normalized:
    # normalizing into 0.1 - 0.9
    X_train_input = normalize((0, 255), (0.1, 0.9), X_train_input)
    X_valid_input = normalize((0, 255), (0.1, 0.9), X_valid_input)
    X_test_input = normalize((0, 255), (0.1, 0.9), X_test_input)
    print("Normalization done.")
    is_normalized = True
    
if (not is_randomized) or always_randomizing:
    X_train_input, y_train_output = shuffle(X_train_input, y_train_output)
    print("Randomizing done.")
    is_randomized = True
    
if join_then_resplit:
    X = np.concatenate([X_train_input, X_valid_input])
    y = np.concatenate([y_train_output, y_valid_output])
    Xt, Xv, yt, yv = train_test_split(X, y, test_size=0.12, random_state=0)
    train_input, train_output = Xt, yt
    valid_input, valid_output = Xv, yv
    test_input, test_output = X_test_input, y_test_output
    print("Train set and valid set joined then resplitted.")
else:
    train_input, train_output = X_train_input, y_train_output
    valid_input, valid_output = X_valid_input, y_valid_output
    test_input, test_output = X_test_input, y_test_output

print("Preprocessing done.")
````

### Model Architecture

````python
### Define your architecture here.
### Feel free to use as many code cells as needed.

# hyper parameters here
learning_rate = 0.1
batch_size = 50
epochs = 150
````

````python
import tensorflow as tf

tf.reset_default_graph()
# deep nn
tf_features = tf.placeholder(tf.float32, [None, n_features])
tf_labels = tf.placeholder(tf.float32, [None, n_labels])

hidden_layers_units = []
layer_tensors = []
num_inputs = n_features
num_outputs = 0
input_node = tf_features
print("Features shape: ", tf_features._shape)
print("Labels shape: ", tf_labels._shape)
for index, num_outputs in enumerate(hidden_layers_units):
    weights = tf.Variable(tf.truncated_normal((num_inputs, num_outputs), 0, 1), dtype=tf.float32, name="DnnWeights_" + str(index))
    biases = tf.Variable(tf.zeros(num_outputs), dtype=tf.float32, name="DnnBiases_" + str(index))
    print("Weights shape for hidden layer {}:".format(index), weights._variable._shape)
    print("Biases shape for hidden_layer {}:".format(index), biases._variable._shape)
    output = tf.add(tf.matmul(input_node, weights), biases)
    activation = tf.nn.relu(output)
    layer_tensors.append((weights, biases, output, activation))
    num_inputs = num_outputs
    input_node = activation
final_index = len(hidden_layers_units)
weights = tf.Variable(tf.truncated_normal((num_inputs, n_labels), 0, 1), dtype=tf.float32, name="DnnWeights_" + str(final_index))
biases = tf.Variable(tf.zeros(n_labels), dtype=tf.float32, name="DnnBiases_" + str(final_index))
logits = tf.add(tf.matmul(input_node, weights), biases)
print("Weights shape for output layer: ", weights._variable._shape)
print("Biases shape for output layer: ", biases._variable._shape)

cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits = logits, labels = tf_labels)
loss = tf.reduce_mean(cross_entropy)
is_correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(tf_labels, 1))
accuracy = tf.reduce_mean(tf.cast(is_correct_prediction, tf.float32))

optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(loss)
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
from tqdm import tqdm
import math

save_file = "./model_deep_gs_{}_{}_{}.ckpt".format(int(learning_rate*1000), batch_size, epochs)
saver = tf.train.Saver()

print("Start training")
print("E: Epoch. L: Last train loss. TA: Last train accuracy. VA: Last valid accuracy")

### Train your model here.
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
### Feel free to use as many code cells as needed.
train_feed_dict = {tf_features: train_in, tf_labels: train_out}
valid_feed_dict = {tf_features: valid_in, tf_labels: valid_out}
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    last_ta, last_l = sess.run([accuracy, loss], feed_dict=train_feed_dict)
    last_va = sess.run(accuracy, feed_dict=valid_feed_dict)
    batches = []
    loss_log = []
    a_train_log = []
    a_valid_log = []
    num_batch = int(math.ceil(len(train_in) / batch_size))
    batch_log_step = 250
    for e in range(epochs):
        for i in tqdm(range(num_batch),unit="batches",
                      desc="E{:03}/{:03}[L:{:>5.2f}|TA:{:.4f}|VA:{:.4f}]".format(e+1, epochs, last_l, last_ta, last_va)):
            batch_start, batch_end = i * batch_size, (i + 1) * batch_size
            batch_in = train_in[batch_start:batch_end]
            batch_out = train_out[batch_start:batch_end]
            _,l = sess.run([optimizer, loss], feed_dict={tf_features: batch_in, tf_labels: batch_out})
            if i % batch_log_step == 0:
                prev_batch = batches[-1] if batches else 0
                batches.append(prev_batch + batch_log_step)
                loss_log.append(l)
                a_train = sess.run(accuracy, feed_dict=train_feed_dict)
                a_valid = sess.run(accuracy, feed_dict=valid_feed_dict)
                a_train_log.append(a_train)
                a_valid_log.append(a_valid)
        
        # accuracy and loss after a full epoch
        last_ta, last_l = sess.run([accuracy, loss], feed_dict=train_feed_dict)
        last_va = sess.run(accuracy, feed_dict=valid_feed_dict)
    
    # final accuracy
    train_accuracy = sess.run(accuracy, feed_dict=train_feed_dict)
    valid_accuracy = sess.run(accuracy, feed_dict=valid_feed_dict)
    
    loss_plt = plt.subplot(211)
    loss_plt.set_title("Loss")
    loss_plt.plot(batches, loss_log, 'g')
    loss_plt.set_xlim(batches[0], batches[-1])
    a_plt = plt.subplot(212)
    a_plt.set_title("Accuracy")
    a_plt.plot(batches, a_train_log, 'r', label='Training Accuracy')
    a_plt.plot(batches, a_valid_log, 'x', label='Valid Accuracy')
    a_plt.set_xlim(batches[0], batches[-1])
    plt.legend(loc=4)
    plt.tight_layout()
    plt.show()
    print("Accuracy on training set: ", train_accuracy)
    print("Accuracy on validation set: ", valid_accuracy)
    
    saver.save(sess, save_file)
    print("Model saved in", save_file)
````
